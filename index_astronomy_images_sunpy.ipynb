## `sunpy`

[`sunpy`](http://sunpy.org/) is an open-source solar data analysis environment for Python.

```python
import matplotlib.pyplot as plt
%matplotlib inline

import sunpy.map
import sunpy.data.sample

aia = sunpy.map.Map(sunpy.data.sample.AIA_171_IMAGE)
```

```python
def plot_image(imagefile):
    fig = plt.figure()
    m = sunpy.map.Map(imagefile)
    ax = plt.subplot(111, projection=m)
    m.plot()
    m.draw_limb()
    m.draw_grid()
    plt.colorbar()

    plt.show()

plot_image(aia)
```

```python
#Example images
from sunpy.data.sample import file_dict as demo_files
image_files = [k for k in demo_files]
print('\n'.join(image_files))
```

```python
def plot_demo_image(k):
    m = sunpy.map.Map(demo_files[k])
    plt.figure()
    m.plot()
    plt.show()
    
plot_demo_image('AIA_193_CUTOUT01_IMAGE')
```

```python
plot_demo_image('AIA_193_IMAGE')
```

```python
#Render with grid, colourbar etc
plot_image(demo_files['AIA_193_IMAGE'])
```
