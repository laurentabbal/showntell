# `bloques` - Control Diagrams

```python
%load_ext tikz_magic
```

```python
%%tikz -p bloques -s 0.5
\renewcommand{\ancholinea}{thin}
\bStart{$X_{ref}$}
\bMinusDown{$X$}
\bGain{$k_{p}$}[line width=0.1mm]
\bPlusUp{$Y$}
\bGainPlus{$U$}{$\omega L$}
\bEnd{$V_{x}$}
```

```python
%%tikz -p bloques -s 0.5
\renewcommand{\ancholinea}{thin}
\bColorB{blue!70!green!20}
\bStart{$X_{r}$}
\bMinusF{NODEX}
\bGain{$k_{p}+\frac{k_{i}}{S}$}
\bFeedBack{$k_{x}$}{NODEX}
\bEnd{$X$}
```

```python
%%tikz -p bloques -s 0.5
\renewcommand{\ancholinea}{very thin}
\bStart{$I_{d(ref)}=0$}
\bMinusDown{$I_{d}$}
\bGain{PI1}
\bPlusF{NODET}
\bEnd{$V_{d}$}
\bNewStart{$\omega_{ref}$}{(-2,-4)}
\bMinusDown{$\omega_{s}$}
\bGain{PI2}
\bMinusUp{$I_{q}$}
\bMarkNodeUp{NODEX}
\bGain{PI3}
\bEnd{$V_{q}$}
\bCrossGain{$\omega L$} {NODEX} {NODET}
```
