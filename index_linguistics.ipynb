## `forest`

## `tikz-dependency`

```python
%%tikz -p tikz-dependency --no-wrap -s 0.5
\begin{dependency}
\begin{deptext}
My \& dog \& also \& likes \& eating \& parsnips \\
\end{deptext}
\depedge{2}{1}{poss}
\depedge{4}{2}{nsubj}
\depedge{4}{3}{advmod}
\depedge{4}{5}{xcomp}
\depedge{5}{6}{dobj}
\end{dependency}
```
